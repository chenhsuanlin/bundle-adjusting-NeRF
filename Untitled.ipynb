```python
def imshow(image_tensor, title=None):
    image_cpu = image_tensor.detach().cpu()
    H, W = image_cpu.shape[-2:]
    plt.imshow(image_cpu.reshape(3, H, W).permute(1,2,0))
    if title != None:
        plt.title(str(title))
    plt.show()
```

```python
loader = tqdm.tqdm(m.train_loader, desc="training epoch {}".format(m.ep+1), leave=False)
for batch in loader:
    var = edict(batch)
    var = util.move_to_device(var, opt.device)
    print(var.image.shape)
    var = m.graph.forward(opt, var, mode='train')
    print(var.keys())
    print(var.image.shape, var.se3_refine.shape, var.ray_idx[:5],
          var.ray_idx.shape, var.rgb.shape, var.depth.shape, var.opacity.shape)
    loss = m.graph.compute_loss(opt, var, mode='train')
    loss = m.summarize_loss(opt, var, loss)
```
